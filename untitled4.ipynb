```python
#Escribiendo nuestra aplicaciГіn
%%writefile mi_app.py
import streamlit as st
import pandas as pd
import plotly.express as px

# Hacemos que la pГЎgina sea mГЎs ancha para que el grГЎfico sea mas grande
st.set_page_config(layout="wide")

st.title("рџ’Ћ Mi mapa 3D de minerГ­a")
st.write("Haz clic sostenido para girar la cueva y usa la rueda del ratГіn para hacer zoom")

# Coordenadas X, Y, Z como en minecraft
datos = pd.DataFrame({
    "Mineral": ["CarbГіn", "Hierro", "Redstone", "Oro", "Esmeralda", "Diamante"],
    "Capa_X": [10, 25, 5, 30, 15, 8],
    "Capa_Z": [15, 5, 20, 10, 25, 2],
    "Profundidad_Y": [50, 35, 15, -10, -30, -55], # El diamante estГЎ en lo mГЎs profundo
    "Experiencia": [10, 20, 30, 40, 50, 100],     # Esto harГЎ que la burbuja sea mГЎs grande
})

# Creando el grГЎfico 3D interactivo
fig = px.scatter_3d(
    datos,
    x='Capa_X',
    y='Capa_Z',
    z='Profundidad_Y',
    color='Mineral',
    size='Experiencia',
    hover_name='Mineral',
    # Le damos los colores reales a los minerales
    color_discrete_map={
        "CarbГіn": "#555555",
        "Hierro": "#D8AF93",
        "Redstone": "#FF0000",
        "Oro": "#FCEE4B",
        "Esmeralda": "#00FF00",
        "Diamante": "#00FFFF"
    }
)

# Apagamos las luces para que parezca una cueva oscura
fig.update_layout(
    template="plotly_dark",
    scene=dict(bgcolor="#111111"),
    margin=dict(l=0, r=0, b=0, t=0)
)

# Mostramos el grГЎfico en Streamlit
st.plotly_chart(fig, use_container_width=True)
```
