```python
import pandas as pd
import sqlite3
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Creamos una conexión a un archivo de base de datos virtual en Colab
conexion = sqlite3.connect("torneo_smash.db")

# 2. Leemos nuestro archivo CSV usando Pandas
df_original = pd.read_csv("smash_dataset_clean.csv", encoding="latin1")

# 3. ¡La Magia! Guardamos toda esa información dentro de una tabla SQL llamada 'combates'
df_original.to_sql("combates", conexion, if_exists="replace", index=False)

print("✅ ¡Éxito total! La base de datos relacional 'torneo_smash.db' ha sido creada y la tabla 'combates' está lista.")
```

```python
# Escribimos nuestra consulta SQL entre tres comillas triples (así podemos usar varias líneas)
# Queremos ver los nombres, los personajes, los KOs y las muertes, ordenados desde el jugador más letal
consulta_sql = """
SELECT Nombre, Personaje, KOs, Muertes
FROM combates
ORDER BY KOs DESC
"""

# Le pedimos a Pandas que ejecute la consulta en nuestra conexión de base de datos
df_filtrado = pd.read_sql_query(consulta_sql, conexion)

# ¡Mostramos el resultado en pantalla!
print("📊 Datos extraídos mediante SQL con éxito:")
df_filtrado.head(10) # head(10) nos muestra los mejores 10 registros
```

```python
# 1. Activamos el estilo de rejilla oscura (Modo Gamer)
sns.set_theme(style="darkgrid")

# 2. Definimos el tamaño de nuestro lienzo (Ancho, Alto) en pulgadas
plt.figure(figsize=(12, 6))

# 3. Creamos el gráfico de barras.
# Sumamos las muertes totales por participante
grafico_barras = sns.barplot(
    data=df_filtrado,
    x="Nombre",
    y="Muertes",
    palette="magma",
    estimator="sum",
    errorbar=None
)

# 4. Añadimos títulos y etiquetas
plt.title("💀 Récord Absoluto de Muertes por Participante", fontsize=16, fontweight="bold", pad=15)
plt.xlabel("Héroes del Aula CLIC", fontsize=12, labelpad=10)
plt.ylabel("Cantidad de Muertes Totales", fontsize=12, labelpad=10)

# 5. Giramos los nombres de los jugadores 45 grados
plt.xticks(rotation=45)

# 6. Mostramos el gráfico
plt.show()
```
